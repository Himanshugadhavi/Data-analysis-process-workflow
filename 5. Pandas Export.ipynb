## to_csv

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
df.groupby('batsman')['batsman_runs'].sum()
```

```python
df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
temp_df.to_csv('batsman_runs.csv')
# we can see the csv file in current working directory
```

```python
# if we don't want the index column in the above csv file, then....
temp_df.to_csv('batsman_runs.csv', index=False)
```

```python
df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs',
              aggfunc='sum')
```

```python
# we cam export this data file to csv file using to_csv
df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs',
              aggfunc='sum').to_csv('batsman_vs_team.csv')
```

## to_excel

```python
df
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_excel('batsman_runs_excel_format.xlsx')
```

```python
# we can save the above data file in particular sheet of the excel file
temp_df.to_excel('batsman_runs_excel_format.xlsx',sheet_name='batsman_runs')
```

```python
# to save the multiple datasheets in the one excel file 
temp_df1 = df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df2 = df.pivot_table(index='batsman', columns='bowling_team', 
                          values='batsman_runs',aggfunc='sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    temp_df1.to_excel(writer, sheet_name='Sheet_name_1')
    temp_df2.to_excel(writer, sheet_name='Sheet_name_2')
```

## to_html

```python
df.head()
```

```python
df.query('batsman_runs == 6').pivot_table(index='over',columns='ball',
                                            values='batsman_runs',
                                            aggfunc='count')

# this will give us the sixes on each ball played in the IPL
```

```python
df.query('batsman_runs == 6').pivot_table(
    index='over',columns='ball',values='batsman_runs',
    aggfunc='count').to_html('sixes_heatmap.html')
```

## to_ json

```python
df.head()
```

```python
df.groupby(['batting_team','batsman'])['batsman_runs'].sum()
```

```python
df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack()
```

```python
df.groupby(['batting_team','batsman'])[
    'batsman_runs'].sum().unstack().to_json('ipl.json')
```
